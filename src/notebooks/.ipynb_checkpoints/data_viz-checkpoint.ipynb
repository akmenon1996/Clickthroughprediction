```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import os 
import seaborn as sns 
import sklearn as skl
from scipy.stats import norm 
import warnings                                         
warnings.filterwarnings('ignore') 
%matplotlib inline 
```

```python
data = pd.read_csv('../../data/watson_marketing.csv')
data.head()
```

```python
# Functions to add data to plot. 
def annotate_plot(ax,w,h,df):                                    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
         ax.annotate(f"{p.get_height() * 100 / df.shape[0]:.2f}%", (p.get_x() + p.get_width() / 2., p.get_height()),
         ha='center', va='center', fontsize=11, color='black', rotation=0, xytext=(0, 10),
         textcoords='offset points')   
            
def annotate_plot_num(ax,w,h,df):                                    
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.1f}'.format(p.get_height()), (p.get_x()+w, p.get_height()+h))
```

```python
%%markdown
## Converting response to a categorical integer. 
```

```python
data.Response = data.Response.apply(lambda x: 0 if x=='No' else 1)
data.head()
```

```python
%%markdown 
# Visualizations 

## Overall Response Rates
```

```python
ax = sns.countplot('Response',data = data)
plt.ylabel('Total number of Response')
annotate_plot(ax, 0.08,1,data)
plt.show()
```

```python
%%markdown

## Stratifying Response rate on variables
```

```python
# Gender
ax = sns.countplot('Response',hue = 'Gender' ,data = data)
plt.ylabel('Total number of Response')
annotate_plot(ax, 0.08,1,data)
plt.show()
```

```python
%%markdown 
### Stratifying by gender does not really reveal too much difference in the responses by males and females. 
```

```python
# Marital Status 
plt.figure(figsize=(12,6))
ax = sns.countplot('Response', hue = data['Marital Status'], data = data)
annotate_plot(ax,0.08,1,data)
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax = sns.countplot('Marital Status', data = data)
annotate_plot(ax,0.08,1,data)
plt.show()
```

```python
%%markdown 
### From the first plot we see that married people make a majority of the people who respond to the marketing campaign. However, from the second plot we see that the number of married people that were reached out to also was far more than single and divorced people. 
```

```python
# Renew Offer

plt.figure(figsize=(8,4))
ax = sns.countplot('Response',hue = 'Renew Offer Type' ,data = data)
plt.ylabel('Total number of Response')
annotate_plot(ax, 0.08,1,data)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax = sns.countplot('Renew Offer Type' ,data = data)
plt.ylabel('Total number of Response')
annotate_plot(ax, 0.08,1,data)
plt.show()
```

```python
%%markdown 
### From the 2 plots we see that the response rate for people with Offer 2 is higher than the rest, even though most people who were targeted were given offer 1. We can now have a fairly strong intuition that offer 2 seems to be performing better than the rest. 

### On the other hand, offer 3 and offer 4 perform really badly with a very low response rate in comparison to the number of people of who recieved the respective offers. 
```
